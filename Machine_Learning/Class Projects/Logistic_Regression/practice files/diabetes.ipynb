```python
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)
iqr = q3-q1
pos_out = q3+1.5*iqr
neg_out = q1-1.5*iqr
```

```python
df = data[~((data>(q3+1.5*iqr))|(data<(q1-1.5*iqr))).any(axis=1)]
```

### Imbalance Dataset

```python
df['Outcome'].value_counts()
```

```python
# majority > 2*minority, so dataset is imbalance
```

```python
x = df.iloc[:,0:8]
```

```python
x.head()
```

```python
y = df.iloc[:,-1]
y.head()
```

```python
from imblearn.over_sampling import RandomOverSampler
ROS = RandomOverSampler()
x_over,y_over = ROS.fit_resample(x,y)
```

```python
print(x_over.shape)
y_over.shape
```

```python
from imblearn.under_sampling import RandomUnderSampler
RUC = RandomUnderSampler()
x_un,y_un = RUC.fit_resample(x,y)
```

```python
print(x_un.shape)
y_un.shape
```

```python
x_un.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_stand = ss.fit_transform(x_un)
```

```python
pd.DataFrame(x_stand)
```

## Split train_test Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_un,y_un,train_size = 0.8, random_state = 50)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Building Logisitc Regression Model

### With UnderSampling Data

```python
from sklearn.linear_model import LogisticRegression
Logit_mod = LogisticRegression()
Logit_mod.fit(x_train,y_train)
```

```python
y_pred_test = Logit_mod.predict(x_test)
```

```python
y_pred_test
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test,y_pred_test)
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
print(accuracy_score(y_test,y_pred_test))
```

### With OverSampling Data

```python
from  sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1 = train_test_split(x_over,y_over,train_size=0.8,random_state=50)
```

```python
print(x_train1.shape)
print(y_train1.shape)
print(x_test1.shape)
print(y_test1.shape)
```
